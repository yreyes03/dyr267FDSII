1. Make a plot of a straight line. Use `linspace()` to create the *x* values and the formula of a straight line, *y = a + b*x*, to create the *y* values (use an *a* and *b* of your choosing). You can pretend x and y are anything you like (x = time, y = international piracy or whatever).

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,15,25)
y = .5*x + 7
plt.plot(x,y);
plt.ylabel('Number of Cells')
plt.xlabel('Time (min)');
```

2. Make some data that are straight line values from the same straight line relationship as in 1. plus random noise. Plot these data.

```python
x = np.linspace(0,15,25)
y = .5*x + 7
n = np.random.randn(25)
#len(n)
plt.scatter(x,y+n);
plt.ylabel('Number of Cells')
plt.xlabel('Time (min)');
```

3. Plot the straight line from 1. and the data from 2. on the same graph. Make sure to add the standard annotations, including a legend.

```python
plt.subplot(2,2,1)
plt.plot(x,y);
plt.ylabel('Number of Cells')
plt.xlabel('Time (min)');

plt.subplot(2,2,2)
plt.scatter(x,y+n);
plt.ylabel('Number of Cells')
plt.xlabel('Time (min)');
```

```python
plt.plot(x,y, label = 'Expected Num. of Cells');
plt.plot(x,y+n, 'bo', label = 'Observed Num. of Cells');
plt.ylabel('Number of Cells')
plt.xlabel('Time (min)');
plt.title('Number of Cells Over Time')
plt.legend();
```

4. Tinker around with your plot (colors, symbols, marker sizes, etc.) until you have a plot you would be happy to use in a presentation. 

```python
plt.plot(x,y, 'r', label = 'Expected Num. of Cells');
plt.plot(x,y+n, 'bo', label = 'Observed Num. of Cells', markersize = 5);
plt.ylabel('Number of Cells')
plt.xlabel('Time (min)');
plt.title('Number of Cells Over Time')
plt.legend();
```

```python
plt.subplot(2,2,1)
plt.plot(x,y);
plt.ylabel('Number of Cells')
plt.xlabel('Time (min)');
plt.title('Expected Number of Cells Over Time')

plt.subplot(2,2,2)
plt.scatter(x,y+n);
plt.ylabel('Number of Cells')
plt.xlabel('Time (min)');
plt.title('Observed Number of Cells Over Time')
plt.tight_layout();
```

5. Make 500 ***normally*** distributed random numbers and make a histogram of them.

```python
nums = np.random.randn(500)
plt.hist(nums, edgecolor = 'k', color = 'r', bins = 15);
```

6. Make 500 ***uniformly*** distributed random numbers (use `...rand()` instead of `...randn()`) and make a histogram of them.

```python
n_nums = np.random.rand(500)
plt.hist(n_nums, color = 'b', edgecolor = 'k', bins = 15);
```

7. Plot the histograms from 5. and 6. in the same axes to compare the two distributions. Tinker around with the `color = ` and `alpha = ` arguments to `plt.hist()` until you're happy with your figure. Don't forget the axis labels and a legend! 

```python
plt.hist(nums, edgecolor = 'k', color = 'r', bins = 15, alpha = .5, label = 'Normal Distribution');
plt.hist(n_nums, color = 'b', edgecolor = 'k', bins = 15, alpha = .3, label = 'Uniform Distribution');
plt.legend()
plt.title('Plotted Random Data Points');
```
