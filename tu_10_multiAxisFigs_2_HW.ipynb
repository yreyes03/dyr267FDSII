In this homework, you'll make a figure containing a scatter plot with marginal histograms similar to what we made earlier. Now, however, we have a few more tools at our disposal so we can probably make a better figure.

Here is some code to make a simulated data consisting of the scores on two tests (A and B) taken by 100 people.

```python
import numpy as np
import matplotlib.pyplot as plt

my_rng = np.random.default_rng(seed = 42)
test_a = my_rng.normal(100, 15, (100,1))
test_b = test_a + my_rng.normal(0, 10, test_a.shape)
```

Let's make a figure featuring a scatter plot of the scores against one another, along with two supporting panels showing the histograms of the two test scores.

An example figure would look something like this:

![hw example figure](images/hw10example.png)

But your goal isn't to make a figure that looks exactly like this; your goal is to make a figure that looks better than this!

```python
dat_dit = {'dat_a': test_a, 'dat_b': test_b}
fig, axd = plt.subplot_mosaic([['dat_s', 'dat_a'], ['dat_s', 'dat_b']], 
                              width_ratios= [2, 1.5], 
                              gridspec_kw = {'hspace': .3, 'wspace': .3, 'top':.85}, 
                              figsize = (8,6))

#plotting test a vs test b scatterplot
axd['dat_s'].scatter(test_a, test_b, color = 'r', alpha = .7);
axd['dat_s'].set_title('Test A Scores vs Test B Scores', size = 13);
axd['dat_s'].set_ylabel('Test B')
axd['dat_s'].set_xlabel('Test A')

#plotting test a histogram
axd['dat_a'].hist(test_a, edgecolor = 'k', color = 'r', bins = 20, alpha = .7);
axd['dat_a'].set_title('Test A Distribution', size = 13);
#axd['dat_a'].set_ylabel('Score')
#axd['dat_a'].set_xlabel('Score', size = 10)
axd['dat_a'].set_xticks([])

#plotting test b histogram
axd['dat_b'].hist(test_b, edgecolor = 'k', color = 'r', bins = 20, alpha = .7);
axd['dat_b'].set_title('Test B Distribution', size = 13);
#axd['dat_b'].set
axd['dat_b'].set_xlabel('Score', size = 10)
axd['dat_b'].set_yticks([0,5,10,15])

fig.set_facecolor('pink')
fig.suptitle('Comparison of Scores for Tests A and B', size = 20);
#plt.tight_layout();
```
